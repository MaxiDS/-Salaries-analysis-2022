```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
results = pd.read_csv("/home/maxi/SoyHenry/stack-overflow-developer-survey-2022/survey_results_public.csv", index_col="ResponseId")
```

```python
#Renombramos las columnas con las que vamos a trabajar, asignandole una variable
languages = "LanguageWantToWorkWith"
salary = "ConvertedCompYearly"
results.rename(columns=({languages:"languages", salary:"salary"}), inplace = True)
results.head()
```

```python
#Eliminamos todos los NaN de las columnas que vamos a trabajar
#Afecta directamente al df, asi que no tenemos que reasignar con "="
results.dropna(subset=["languages","salary"], inplace=True)
```

```python
#Seleccionamos filas y columnas con las que vamos a trabajar formato: loc[fila,columna]

results = results.loc[:,["languages","salary"]].sort_values(by="salary")

#vemos que hay numeros muy inreales de salarios como "1" o "2e6"
results
```

```python
#realizamos un ajuste con valores mas reales:

ajuste = (results["salary"] >= 6000) & (results["salary"] <= 1e6)
results= results[ajuste]
```

```python
results
```

```python
#Ahora tenemos que limpiar los lenguajes 
ajuste_language = results["languages"]
#creamos una lista vacia, en este caso seleccione un index al azar y aplico split para dar un ejemplo de como tiene que quedar dentro
allLanguages = ajuste_language[28010].split(";")
allLanguages
```

```python
#Creamos nuestro loop para ir seleccionando los lengujes
# row = fila (english) xd 

for row in ajuste_language:
    for leng in row.split(";"):
        allLanguages.append(leng)

#limpiamos nuestra variable para que no existan repetidos

allLanguages = list(set(allLanguages))

allLanguages
```

```python
Meansalary = pd.DataFrame(data=np.zeros(len(allLanguages)), index= allLanguages, columns=["Salary"])
Meansalary
```

```python
#Importa libreria "re" para sacar el caracter especial de C++
import re

for leng in allLanguages:
    mask = results["languages"].str.contains(re.escape(leng))
    average = results.loc[mask,"salary"].mean()
    Meansalary.loc[leng,"Salary"] = average


Meansalary = Meansalary.sort_values(by="Salary")
Meansalary
```

```python
    #Graficar
    #Colores
    colors = [
        "#03071E",
        "#370617",
        "#6A040F",
        "#9D0208",
        "#D00000",
        "#DC2F02",
        "#E85D04",
        "#F48C06",
        "#FAA307",
        "#FFBA08"
    ]

    fontdict = {
        'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
    }
```

```python
%matplotlib qt5

plt.style.use("seaborn")
plt.barh(Meansalary.index, Meansalary["Salary"], color = colors)
plt.ylabel("LENGUAGES")
plt.xlabel("SALARY")
plt.title("MEAN SALARY 2022")
plt.tight_layout()

# vline(valor a tomar , comienza , termina)
plt.vlines(Meansalary["Salary"].mean(),0, len(allLanguages) ,color = "red" , linestyles= "dashed", label= "MeanSalary" )
plt.legend()

#poner $ a las x

plt.gca().xaxis.set_major_formatter("${x:1.0f}")



plt.show()
```
